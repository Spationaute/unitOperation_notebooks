Example 2.7 

n-heptane/toluene system at $\textbf{1}$ $\textbf{atm}$

Regular solution theory $\textbf{without}$ Flory–Huggins correction:

$$\phi_i=\frac{x_i \nu_{iL}}{\nu_L}$$
$$ ln \gamma_{iL}=\frac{\nu_{iL} {\bigg( \delta_i-\sum_{j=1}^{C} \phi_j \delta_j \bigg)}^2}{RT} $$

Regular solution theory $\textbf{with}$ Flory–Huggins correction:

$$\phi_i=\frac{x_i \nu_{iL}}{\nu_L}$$
$$ ln \gamma_{iL}=\frac{\nu_{iL} {\bigg( \delta_i-\sum_{j=1}^{C} \phi_j \delta_j \bigg)}^2}{RT}+ln \bigg(\frac{\nu_{iL}}{\nu_L}\bigg)+1-\frac{\nu_{iL}}{\nu_L}$$


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
H = 'n-heptane'
T = 'toluene'
P = 1 
```

```python
def gamma_H(x,T,sp_H,sp_T,nu_HL,nu_TL):
    phi_H=x*nu_HL/(x*(nu_HL)+(1-x)*(nu_TL))
    phi_T=1-phi_H
    t1=nu_HL*((sp_H-phi_H*sp_H-phi_T*sp_T)**2)
    t2=t1/(R*T)
    ans=np.exp(t2)
    return float(ans)

def gamma_T(x,T,sp_H,sp_T,nu_HL,nu_TL):
    phi_T=x*nu_TL/(x*(nu_TL)+(1-x)*(nu_HL))
    phi_H=1-phi_T
    t1=nu_TL*((sp_T-phi_T*sp_T-phi_H*sp_H)**2)
    t2=t1/(R*T)
    ans=np.exp(t2)
    return float(ans)

def gamma_H_FH(x,T,sp_H,sp_T,nu_HL,nu_TL):
    phi_H=x*nu_HL/(x*(nu_HL)+(1-x)*(nu_TL))
    phi_T=1-phi_H
    t1=nu_HL*((sp_H-phi_H*sp_H-phi_T*sp_T)**2)/(R*T)
    t2=np.log(nu_HL/(x*nu_HL+(1-x)*nu_TL))+1-nu_HL/(x*nu_HL+(1-x)*nu_TL)
    ans=np.exp(t1+t2)
    return float(ans)

def gamma_T_FH(x,T,sp_H,sp_T,nu_HL,nu_TL):
    phi_T=x*nu_TL/(x*(nu_TL)+(1-x)*(nu_HL))
    phi_H=1-phi_T
    t1=(nu_TL*(sp_T-phi_T*sp_T-phi_H*sp_H)**2)/(R*T)
    t2=np.log(nu_TL/(x*nu_TL+(1-x)*nu_HL))+1-nu_TL/(x*nu_TL+(1-x)*nu_HL)
    ans=np.exp(t1+t2)
    return float(ans)


xT=np.zeros(7)
xH=np.array([1.000,0.9154,0.7479,0.5096,0.2681,0.1087,0]) #liquid phase composition data
yH=np.zeros(7)
yT=np.zeros(7)
yH_FH=np.zeros(7)
yT_FH=np.zeros(7)

R=1.987            # gas constant

xT = 1-xH


Tc=np.array([98.41, 98.70, 99.58, 101.47,104.52,107.57,110.6])  #temperature data

Tk = Tc+273.15   #kelvin

sp_H = 7.43     #solubility parameter (H) 
sp_T = 8.914    #solubility parameter (T)

nu_HL = 147.5    #liquid molar volume (H)
nu_TL= 106.8     #liquid molar volume (T)

```

```python
for i in range(7):
    yH[i]=gamma_H(xH[i],Tk[i],sp_H,sp_T,nu_HL,nu_TL)
    yT[i]=gamma_T(xT[i],Tk[i],sp_H,sp_T,nu_HL,nu_TL)
    yH_FH[i]=gamma_H_FH(xH[i],Tk[i],sp_H,sp_T,nu_HL,nu_TL)
    yT_FH[i]=gamma_T_FH(xT[i],Tk[i],sp_H,sp_T,nu_HL,nu_TL)
    
```

```python
plt.rcParams["figure.figsize"] = 8,9
plt.plot(xH, yH ,xH, yT)
plt.title('Liquid-phase activity coeffcient for {:s}-{:s} system at {:.1f} [atm]'.format(H,T,P))
plt.legend(['n-heptane','toluene'],loc='best')
plt.ylabel('Liquid-phase activity coefficient')
plt.xlabel('mole fraction of {:s}'.format(H))
plt.xlim(0,1)
plt.ylim(1,1.7)
plt.grid();
```

```python
plt.rcParams["figure.figsize"] = 8,9
plt.plot(xH, yH_FH,'--', xH,yT_FH,'--')
plt.title('Liquid-phase activity coeffcient for {:s}-{:s} system at {:.1f} [atm]'.format(H,T,P))
plt.legend(['n-heptane','toluene'],loc='best')
plt.ylabel('Liquid-phase activity coefficient')
plt.xlabel('mole fraction of {:s}'.format(H))
plt.xlim(0,1)
plt.ylim(1,1.7)
plt.grid();
```
