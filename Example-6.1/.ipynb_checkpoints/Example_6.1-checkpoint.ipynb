For dilute ethyl alchohol, the K-vakue is determined from a modified Rault's law, 

$$K=\frac{\gamma P^{s}}{P}$$

The vapour pressure at 30'C is 10.5 Kpa, and from infinite dilution in water data at 30 C, the liquid-phase activity coefficient of ethyl alchohol is 6.

$$K=\frac{6 \times 10.5}{110}=0.57$$

The minimum solute-free absorbent rate is given by:

$$L_{min}'=V' K_N$$

where $V'$ is solute-free gas rate:

$$0.98 \times 180 = 176.4 \ \ kmol/h $$

Thus, the minimum abosrbent rate for $97.5 \%$ recovery is:

$$L_{min}'=176.6 \times 0.57 \times 97.5 \ \ kmol/h$$

The solute-free absorbent rate at $50 \%$ above the minimum is:

$$L'=1,5 \times 97.5=146.2 \ \ kmol/h$$

The alchohol recovery of $97 \%$ corresponds to:

$$0.97 \times 0.02 \times 180= 3.49 \ \ kmol/h$$

The amount of ethyl alcohol remaining in the exiting gas is:

$$(1-0.97)(0.02)(180)=0.11 \ \ kmol/h$$

Top: $$x_0=0 \ \ \; \ \ \ Y_1=\frac{0.11}{176.4}=0.0006$$

Bottom: $$Y_{N+1}=\frac{0.11+3.49}{176.4}=0.0204 \ \ \; \ \ \ x_N=\frac{3.49}{146.2}=0.0239 $$

$$Y_{n+1}=X_n (\frac{L'}{V'})+Y_1-X_0(\frac{L'}{V'}) $$

$$Y_{N+1}=\frac{146.2}{176.4}x_N+0.0006=0.829X_N+0.0006$$

This a dilute system. The equilibrium curve using K=0.57:

$$K_n=\frac{y_n}{x_n}=\frac{Y_n/(1+Y_n)}{X_n/(1+X_n)}$$

$$0.57=\frac{Y/(1+Y)}{X/(1+X)}$$

$$Y=\frac{0.57X}{1+0.43X}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
%matplotlib inline

X = np.linspace(0,0.04)

Y_e = lambda X: (0.57*X)/(1+0.43*X)

Y_o = lambda X: 0.829*X+0.0006

plt.figure(figsize=(10,6))

plt.plot(X, [Y_e(X) for X in X],X, [Y_o(X) for X in X])

plt.xlabel('Moles of alchohol/mole of alchohol-free liquid, X')
plt.ylabel('Moles of alchohol/mole of alchohol-free gas, Y')
plt.legend(['Equilibrium curve','Operating line'])
plt.ylim(0,0.035)
plt.xlim(-0.005,0.04)
plt.grid()

x0=0
yP=0.0006

nTray = 0

xN=0.0239
YN_1=0.0204

plt.plot(xN,YN_1,'bo',ms=7)
plt.text(xN-0.005,YN_1+0.001,"Y_N+1, X_N")
plt.plot(x0,yP,'bo',ms=7)
plt.text(x0-0.002,yP+0.001,"Y1, X0")

while x0 < 0.0239:
    nTray += 1

    Tdew = fsolve(lambda X:Y_e(X) - yP, 0.0005)
    x2 = Tdew
    y2 = yP

    x1=x0
    y1=yP

    x3=x2
    y3=Y_o(x2)

    plt.plot([x1,x2],[y1,y2],'r')
    plt.plot([x2,x3],[y2,y3],'r')

    plt.plot(x2,y2,'ro',ms=7)
    
    plt.text(x2+0.0005,y2-0.001,nTray)

    yP=y3
    x0=x3

   
plt.text(0.001,0.01,'nTrays = {:d}'.format(int(nTray)))
```
