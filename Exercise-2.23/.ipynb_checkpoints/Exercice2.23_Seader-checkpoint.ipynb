ethanol/benzene system at $\textbf{145}$ $\textbf{C}$

Wilson equation (species 1):

$$ ln \gamma_{1}= -ln (x_1+\Lambda_{12} x_2)+x_2 \bigg( \frac{\Lambda_{12}}{x_1+\Lambda_{12} x_2}-\frac{\Lambda_{21}}{x_2+\Lambda_{21} x_1} \bigg) $$

Wilson equation (species 2):

$$ ln \gamma_{2}= -ln (x_2+\Lambda_{21} x_1)-x_1 \bigg( \frac{\Lambda_{12}}{x_1+\Lambda_{12} x_2}-\frac{\Lambda_{21}}{x_2+\Lambda_{21} x_1} \bigg) $$


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
E = 'ethanol'  #1
B = 'benzene'   #2
P = 1 
T= 45 #centigrade 
```

```python
def gamma_1(x1,lam12,lamb21):
    
    x2=1-x1
    t1=-1*np.log(x1+lamb12*x2)
    t2=lamb12/(x1+lamb12*x2)
    t3=lamb21/(x2+lamb21*x1)
    ans=t1+x2*(t2-t3)
    return ans



def gamma_2(x2,lam12,lamb21):
    
    x1=1-x2
    t1=-1*np.log(x2+lamb21*x1)
    t2=lamb12/(x1+lamb12*x2)
    t3=lamb21/(x2+lamb21*x1)
    ans=t1-x1*(t2-t3)
    return ans



xE =np.linspace(0, 1, num=100)

xB=1-xE

yE=np.zeros(xE.size)
yB=np.zeros(xB.size)

R=1.987            # gas constant

lamb12=0.124
lamb21=0.523
```

```python
for i in range(xE.size):
    yE[i]=gamma_1(xE[i],lamb12,lamb21)
    yB[i]=gamma_2(xB[i],lamb12,lamb21)


    
```

```python
plt.rcParams["figure.figsize"] = 8,9
plt.plot(xE, yE ,xE, yB)
plt.title('Liquid-phase activity coeffcient for {:s}-{:s} system at {:.1f} [atm]'.format(E,B,P))
plt.legend(['Ethnaol','Benzene'],loc='best')
plt.ylabel('Liquid-phase activity coefficient')
plt.xlabel('mole fraction of {:s}'.format(E))
plt.xlim(0,1)
plt.ylim(0,3)
plt.grid();
```
